```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/creditcard.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
fraud.shape
```

```python
legit_sample = legit.sample(n=492)
```

```python
df = pd.concat([legit_sample,fraud],axis=0)
```

```python
df['Class'].value_counts()
```

```python
x = df.drop(columns='Class',axis=1)
y = df['Class']
```

```python
x.shape
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
```

```python
lr = LogisticRegression(max_iter =1000)
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(y_train)
```

```python
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
train_acc = accuracy_score(y_train,lr.predict(x_train))
test_acc = accuracy_score(y_test,lr.predict(x_test))

print(f"Training Accuracy: {train_acc}")
print(f"Test Accuracy: {test_acc}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,x_test)
```

```python
y_pred_rfc = rfc.predict(y_train)
```

```python
acc_rfc = accuracy_score(y_test,y_pred_rfc)
```

```python
acc_rfc
```

```python
train_acc = accuracy_score(y_train,rfc.predict(x_train))
test_acc = accuracy_score(y_test,rfc.predict(x_test))
```
